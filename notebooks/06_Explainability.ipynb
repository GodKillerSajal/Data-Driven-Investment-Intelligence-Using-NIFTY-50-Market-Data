# 06️⃣ Explainable AI Framework

Use SHAP values to explain model attributions for the stationary features model.

```python
import os, pathlib, pandas as pd, joblib, matplotlib.pyplot as plt, shap
import matplotlib
matplotlib.use('Agg')
```

```python
NOTEBOOK_DIR = pathlib.Path(os.path.abspath('') if '__file__' not in locals() else os.path.dirname(__file__))
PROJECT_ROOT = NOTEBOOK_DIR.parent
DATA_PARQUET = PROJECT_ROOT / 'data' / 'parquet'
MODEL_DIR = PROJECT_ROOT / 'models'
OUTPUT_DIR = PROJECT_ROOT / 'outputs'

model = joblib.load(MODEL_DIR / 'gbr_model.pkl')
df = pd.read_parquet(DATA_PARQUET / 'modeled_data.parquet')
df = df.dropna()

features = [
    'Return', 'MA20_Ratio', 'MA50_Ratio', 'EMA20_Ratio',
    'Volatility', 'Momentum_Ratio', 'RSI', 'MACD_Ratio',
    'Signal_Ratio', 'BB_Width', 'BB_Pos'
]
X = df[features]

# Sample 500 points for SHAP computation
X_sample = X.sample(min(500, len(X)), random_state=42)

explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(X_sample)

# Generate and save plot
plt.figure(figsize=(10, 6))
shap.summary_plot(shap_values, X_sample, show=False)
plt.title('SHAP Feature Importance Summary (Stationary Model)')
plt.tight_layout()
plt.savefig(OUTPUT_DIR / 'shap_summary_plot.png')
plt.close()
print('SHAP analysis complete.')
```
